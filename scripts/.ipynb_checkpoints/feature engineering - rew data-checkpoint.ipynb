```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('mode.chained_assignment', None)

from sklearn import ensemble
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
import seaborn as sns
```

```python
df = pd.read_csv('property_listing.csv')
sdf = pd.read_csv('School_Data_with_Address.csv')
crdf = pd.read_csv('crimewithpost.csv')

# Remove Duplicates
cdf = df.drop_duplicates(['listing_id'])
```

```python
cdf.head()
```

```python
# Remove null fields
cdf = cdf[cdf['postal_code'].notnull()]

# Get area from postal code
cdf['area'] = cdf['postal_code'].str.split(expand=True)[0]

# Keeping only Vancouver areas
cdf = cdf[cdf['area'].str.startswith('V')]

# Clear NaN's in fireplaces field
cdf['fireplaces'] = cdf['fireplaces'].fillna(0)

# Clean age field
cdf['age'] = cdf['age'].str.extract(r'\((\d+)', expand=True)
cdf['age'] = cdf['age'].fillna(0)
cdf['age'] = cdf['age'].astype(int)
```

```python
 
# Clean price, bath, bed and area_sqft fields
for i in cdf.index:
    cdf.loc[i,'price'] = float(cdf.loc[i,'price'][1:].replace(",", ""))
    
cdf['bath'] = cdf['bath'].fillna(0)
cdf['bed'] = cdf['bed'].fillna(0)
cdf['area_sqft'] = cdf['area_sqft'].fillna(0)
```

```python
cdf.to_csv('clean_rew_data.csv')
cdf.head()
```

```python
# Integrate School data with REW data

ar = cdf['area'].value_counts()
l1 = ar.index

sar = sdf['postal_area'].value_counts()
l2 = sar.index

maxs = sdf.groupby('postal_area')['2017-18 rating'].max()
avgs = sdf.groupby('postal_area')['2017-18 rating'].mean()
scs = sdf.groupby('postal_area')['postal_area'].count()

avgsr = {}
maxsr = {}
sc = {}

for i in l1:
    if i not in l2:
        avgsr[i] = 0
        maxsr[i] = 0
        sc[i] = 0
    else:
        avgsr[i] = avgs.loc[i]
        maxsr[i] = maxs.loc[i]
        sc[i] = scs.loc[i]

        
cdf['num_schools'] = 0
cdf['avg_school_rating'] = 0
cdf['max_school_rating'] = 0

cdf['num_schools'] = cdf['area'].map(sc)
cdf['avg_school_rating'] = cdf['area'].map(avgsr)
cdf['max_school_rating'] = cdf['area'].map(maxsr)
```

```python
def get_vals(st):
    dt1 = cdf[cdf[st]>0]
    x = dt1.groupby('area')['price'].mean()
    y = dt1.groupby('area')[st].mean()
    mp = x.to_dict()
    rp = y.to_dict()

    dic ={}
    for k in rp.keys():
        dic[mp[k]/1000000] = rp[k]
    return dic

dt1 = get_vals('num_schools')
plt.figure(figsize=(25,8))
plt.subplot(1, 3, 1)
plt.scatter(dt1.keys(),dt1.values())
plt.xlabel('Avg. Property price (million)',fontsize=15)
plt.ylabel('Number of schools',fontsize=15)

dt2 = get_vals('avg_school_rating')
plt.subplot(1, 3, 2)
plt.scatter(dt2.keys(),dt2.values())
plt.xlabel('Avg. Property price (million)',fontsize=15)
plt.ylabel('Average school rating',fontsize=15)

dt3 = get_vals('max_school_rating')
plt.subplot(1, 3, 3)
plt.scatter(dt3.keys(),dt3.values())
plt.xlabel('Avg. Property price (million)',fontsize=15)
plt.ylabel('Maximum School rating',fontsize=15)

plt.suptitle('Correlation between school data and property prices',fontsize=15)
plt.show()
```

```python
# Integrate Crime data with REW data

ar = cdf['area'].value_counts()
arl = ar.index

x = crdf['Postal'].value_counts()
cl = x.index

cp ={}
totalcrimes = float(x.values.sum())
for i in x.index:
    p = (x.loc[i]/totalcrimes)*100
    cp[i] = p
    
for i in arl:
    if i in cp.keys():
        continue
    cp[i] = 0
    
cdf['crime_percentage'] = 0
cdf['crime_percentage'] = cdf['area'].map(cp)
```

```python
# OUTLIER DETECTION

medprcs = cdf.groupby(['area'])['price'].median().to_dict()

for i in medprcs.keys():
    
    temp = cdf[cdf['area']==i]
    l=[]

    for j in temp.index:
        l.append(abs(temp.loc[j,'price']-medprcs[i]))
    
    m = np.median(l)
    s = medprcs[i]-(2*m)
    e = medprcs[i]+(2*m)
    t= temp[(temp['price']>=s) & (temp['price']<=e)]

    for k in temp.index:
        if k not in t.index:
            cdf = cdf[cdf.listing_id != temp.loc[k,'listing_id']]
```
