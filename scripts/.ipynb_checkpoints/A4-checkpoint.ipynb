# Question 8. PDF and CDF
Can you believe that you have already drawn 8 interesting findings by exploring a single column! This is the power of EDA combined with critical thinking. Now we are moving to multivariate analysis.

Suppose you want to compare the housing price between this year and last year, i.e., CURRENT_PRICE vs. PREVIOUS_PRICE. You can plot their distributions, and make the comparison. There are two ways to define a distribution: Probabilistic Distribution Function (PDF) and Cumulative Distribution Function (CDF).

In the following, please make two plots and put them side-by-side.

In the first plot, use histograms to plot the probabilistic distributions of CURRENT_PRICE and PREVIOUS_PRICE.
In the second plot, use histograms to plot the cumulative distributions of CURRENT_PRICE and PREVIOUS_PRICE.
There are a few properties which are way more expensive than the others. For both plots, please exclude those properties by setting xlim = (0, 5Million).

```python
df1 = df[(df['CURRENT_PRICE'] < 5000000)]['CURRENT_PRICE']/1000000
df2 = df[(df['PREVIOUS_PRICE'] < 5000000)]['PREVIOUS_PRICE']/1000000
 
```

```python
fig, axs = plt.subplots(figsize = (18,5),nrows=1, ncols=2, sharex=True, tight_layout=True)
# Probabilistic Distribution Function (PDF)
axs[0].hist(df1, bins = 500, alpha = 0.5)
axs[0].hist(df2, bins = 500, alpha = 0.5)
axs[0].set(xlabel="Property Price (Million)", ylabel='No. of properties')
axs[0].set_title('Probabilistic Distribution Function (PDF)')
axs[0].set_xlim(0,5)
axs[0].legend(loc = 0)
axs[0].grid()
#Cumulative Distribution Function (CDF)
axs[1].hist(df1,cumulative = True,density = True, bins = 500, alpha = 0.5)
axs[1].hist(df2,cumulative = True,density = True, bins = 500, alpha = 0.5)
axs[1].set(xlabel="Property Price (Million)", ylabel='Probability')
axs[1].set_title('Cumulative Distribution Function (CDF)')
axs[1].set_xlim(0,5)
axs[1].grid()
axs[1].legend(loc = 0)
```

# Findings:

1. The shift of the current price in the left plot suggests how the price rose from 0.5 to 1 m.
2. The plot on the right suggests a steady trend in the cumulative distributions of all the prices started from 1.5 millon and above.

# Question 10. Use EDA to answer interesting questions (2)
Although the housing price of the entire Vancouver area is increasing, there might be some areas whose housing price is decreasing. To answer this question, we need to consider another column -- PROPERTY_POSTAL_CODE.

PROPERTY_POSTAL_CODE (e.g., "V5A 1S6") is a six-character string with a space separating the third and fourth characters. We use the first three characters to represent an area.

We first filter out the areas which have less than 10 properties. For each of the remaining areas, we calculate the percentage of the properties whose price has decreased compared to the last year. For example, if an area "V5A" has 50 properties, and 30 of them have decreased, then the percentage is 60%.

Please write code to find the top-10 areas with the highest percentages. Create a bar chart to visualize them.

```python
new = df["PROPERTY_POSTAL_CODE"].str.split(" ", n = 1, expand = True) 
df["AREA"]= new[0] 
df["SUB_AREA"]= new[1] 

grouped_areas = df.groupby('AREA').AREA.agg('count').to_frame('total_count').reset_index() 
grouped_areas = grouped_areas[grouped_areas['total_count'] > 10] 
merged = df.merge(grouped_areas, how='right', indicator=True)
merged = merged[merged['PREVIOUS_PRICE'] - merged['CURRENT_PRICE'] > 0]
decreased = merged.groupby('AREA').AREA.agg('count').to_frame('decreased_count').reset_index() 
final = decreased.merge(grouped_areas,on = 'AREA',how='right', indicator='exists')
final = final.dropna()
final['per_decrease']= (final['decreased_count']/final['total_count'])*100
final = final.sort_values(['per_decrease'],ascending=False).reset_index(drop=True).iloc[0:10 : ]
```

```python
fig, ax = plt.subplots(figsize = (18,5)) 
ax = sns.barplot(final.AREA, final.per_decrease, ax = ax,order = final.AREA, alpha = 0.5) 
ax.grid()
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax.set(xlabel="AREA", ylabel='Percentage (%)')
ax.set_xticklabels(final.AREA)
plt.ylim(0,100,20)
ax.set_title('Top 10 areas where the % of decrese in property count is higher')
for i, v in enumerate(final["per_decrease"].iteritems()):        
    ax.text(i ,v[1], "{:,}".format(v[1]), color='m', va ='bottom', rotation=45)
```

# Findings:

    1. Area code with V6S and V6L experianced the highest decrease in price.
    2. Overall there were similar drop in price among areas like V6N, V6R, V5W and V5P.

# Part 2. Bootstrapping
In Part 1, we run our analysis over the full dataset. In reality, however, you may not be that lucky. It is more often than not that you can only collect a sample of the data. Whenever you derive a conclusion from a sample (e.g., The Vancouver's housing price has increased by 10\% since last year), you should ALWAYS ask yourself: "CAN I TRUST IT?". In other words, you want to know that if the same analysis was conducted on the full data, would the same conclusion be derived? In Part 2, you will learn how to use bootstrapping to answer this question.

Please download the sample dataset property_tax_report_2018_sample.zip, and load it as a DataFrame.

```python
df_sample = pd.read_csv("property_tax_report_sample.csv")

df_sample['CURRENT_PRICE'] = df_sample.apply(lambda x: x['CURRENT_LAND_VALUE']+x['CURRENT_IMPROVEMENT_VALUE'], axis = 1)

df_sample['PREVIOUS_PRICE'] = df_sample.apply(lambda x: x['PREVIOUS_LAND_VALUE']+x['PREVIOUS_IMPROVEMENT_VALUE'], axis = 1)

df_sample = df_sample[df_sample['LEGAL_TYPE'] == 'STRATA']
```

# Task 1. Analysis Result Without Bootstrapping
Please compute the median of PREVIOUS_PRICE and CURRENT_PRICE, respectively, and compare them in a bar chart.

```python
pp_med = df_sample['PREVIOUS_PRICE'].median()/1000000
cu_med = df_sample['CURRENT_PRICE'].median()/1000000
print('medians ', cu_med, 'and', pp_med)
sns.barplot(['Previous price', 'Current price'],[pp_med, cu_med],alpha = 0.5)
plt.ylim(0,1)
```

# Task 2. Analysis Result With Bootstrapping
From the above chart, we find that the median of PREVIOUS_PRICE is about 0.6 M, and the median of CURRENT_PRICE is about 0.7 M. Since the numbers were obtained from the sample, "CAN WE TRUST THESE NUMBERS?"

In the following, please implement the bootstrap by yourself, and compute a 95%-confidence interval for each number. This document gives a good tutorial about the bootstrap. You can find the description of the algorithm in Section 7.



```python
bootstrap_sample = df_sample[['PREVIOUS_PRICE','CURRENT_PRICE']]

current_delta = []
previous_delta = []
for i in range(1000):
        sample = bootstrap_sample.sample(n = len(bootstrap_sample), replace = True)
        current_delta.append(cu_med - (sample.CURRENT_PRICE.median()/1000000))
        previous_delta.append(pp_med - (sample.PREVIOUS_PRICE.median()/1000000))

current_delta.sort()
previous_delta.sort()

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), math.fabs(np.std(a)/math.sqrt(n))
    h = se * ((1 + confidence) / 2)
    return m, m-h, m+h

c_mean,c_lower,c_upper = mean_confidence_interval(current_delta)
p_mean,p_lower,p_upper = mean_confidence_interval(previous_delta)

ax=sns.barplot(['Previous price', 'Current price'],[pp_med, cu_med],alpha = 0.5, yerr = [[p_lower,c_lower],[p_upper, c_upper]])
plt.ylim(0,1)
plt.grid()
```
