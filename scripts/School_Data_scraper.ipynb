```python
import requests
import pandas as pd
import lxml.html
from lxml.html import fromstring
import time
```

```python
school_urls = []
i=1

filesnames = ['1.html','2.html','3.html','4.html']

html_txt = ''

for name in filesnames:
    with open('Schools_pages/'+name,'r',encoding="utf-8") as f:
        html_txt+=f.read()

Address_info = []

sch_data = pd.read_csv('Schools_Data.csv',encoding='latin1')

for val in list(sch_data['School Name']):
    tree = lxml.html.fromstring(html_txt)
    link = tree.xpath('//a[contains(text(),\"'+val+'\")]/@href')
    response = requests.get(link[0])
    tree = lxml.html.fromstring(response.text)
    school_info = tree.xpath("//span[@id='ctl00_ContentPlaceHolder1_SchoolInfoDisplay']/text()")
    info=''
    for i in school_info:
        info+=i+' '
    Address_info.append(info)

sch_data['Address_info'] = Address_info
```

```python
sch_data = sch_data.drop_duplicates(subset='School Name')
```

```python
import re

ar_code = []
sch_data['Address_info'] = sch_data['Address_info'].fillna('unknown')

for val in sch_data['Address_info']:
    if re.search("[V][0-9][ABCEGHJKLMNPRSTVWXYZ]", val , re.IGNORECASE):
        ar_code.append(re.search("[V][0-9][ABCEGHJKLMNPRSTVWXYZ]", val , re.IGNORECASE | re.DOTALL)[0])
    else:
        ar_code.append("none")

sch_data["postal_area"] = ar_code
```

```python
import re

p_code = []
sch_data['Address_info'] = sch_data['Address_info'].fillna('unknown')

for val in sch_data['Address_info']:
    if re.search("[V][0-9][ABCEGHJKLMNPRSTVWXYZ]", val , re.IGNORECASE):
        p_code.append(re.search(r'[A-Z]\d[A-Z]\s\d[A-Z]\d', val , re.IGNORECASE | re.DOTALL)[0])
    else:
        p_code.append("none")

sch_data["postal_code"] = p_code
```
